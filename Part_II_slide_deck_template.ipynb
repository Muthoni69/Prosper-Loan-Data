## Investigation Overview
In this investigation, I wanted to look at how an individual's Monthly Income affect the amount of loan given and if the Monthly Loan Payments depend on the Income. The main focus was OriginalLoan, Monthly Income, Monthly Loan Payments, Loan Status, Employment Status and the Credit Scores.


## Dataset Overview

The prosperLoanData dataset contains 113,937 loans with 81 variables on each loan, including loan amount,borrower rate, current loan status, borrower income, Interest and Fees rates and many others. Some data points were removed due to the presence of missing values.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('prosperLoanData.csv')
```

```python
# Data wrangling, melting the datset, removing all missing values , Rename some columns that have long column names to make the dataset more tidy
df1 = df.filter(items=['ListingNumber', 'Term', 'LoanStatus', 'BorrowerRate', 'ProsperScore', 'ListingCategory', 'EmploymentStatus', 'EmploymentStatusDuration', 'IsBorrowerHomeowner', 'CreditScoreRangeLower', 'StatedMonthlyIncome', 'TotalProsperLoans', 'LoanMonthsSinceOrigination', 'LoanOriginalAmount', 'MonthlyLoanPayment', 'TotalProsperPaymentsBilled', 'LP_CustomerPayments', 'LP_GrossPrincipalLoss'])
df1.rename(columns={'BorrowerRate':'InterestRate', 'EmploymentStatusDuration':'EmploymentDuration', 'IsBorrowerHomeOwner':'IsHomeOwner','CreditScoreRangeLower':'CreditScore', 'StatedMonthlyIncome':'MonthlyIncome', 'TotalProsperLoans':'ProsperLoans', 'LoanMonthsSinceOrigination':'LoanStretch', 'LoanOriginalAmount':'OriginalLoan', 'MonthlyLoanPayment':'MonthlyPayment', 'TotalProsperPaymentsBilled':'TotalProsperPaid'}, inplace=True)
df_clean = df1.copy()
df_clean = df_clean.dropna()
```

## Monthly Income and Monthly Loan Payments 

The two graphs are hightly skewed to the right and the peak points in both the graphs are almost similar. This shows that there's a relationship between the monthly income of an individual and the monthly loan payments made.

```python
plt.figure(figsize=[20,5])

plt.subplot(1,2,1)
bins = np.arange(0, df_clean['MonthlyIncome'].max()+450, 450)
plt.xlim(0, 36000)
plt.xlabel('Monthly Income distribution')
plt.hist(data=df_clean, x= 'MonthlyIncome', bins=bins);

plt.subplot(1,2,2)
bins = np.arange(0, df_clean['MonthlyPayment'].max()+10, 10)
plt.xlim(0, 1000)
plt.xlabel('Monthly Payments')
plt.hist(data=df_clean, x='MonthlyPayment', bins=bins);

plt.suptitle('Monthly Income and Monthly Payments vs Original Loans');
```

## Loan Status and Employment Status 
The plot suggests that employed borrowers have the highest current and completed loans. The largest number of charged off loans also belong to the employed/self-employed borrowers. Full-time employed borrowers who are also employed have a significant amount of loans with Completed being the highest. Unemployed, retired and part-time employed borrowers have the lowest amount of loans.

```python
plt.figure(figsize=[10,6])

ax = sb.countplot(data=df_clean, x ='EmploymentStatus', hue = 'LoanStatus')
ax.legend(loc=1, framealpha=1)
plt.xticks(rotation=15);
plt.title('Loan Status vs  Employment Status');
```

## Credit Scores, Monthly Payments and Employment Status, Loan Status

Surprisingly, it appears that the unemployed borrowers have a higher credit score compared to the employed even though it's the employed who make larger monthly loan payments. Also, the retired category have a higher credit score, this is understable because the credit may have accumulated before retirement. Additionally, borrowers who are on their final loan payment processes are the ones with the highest credit scores. The borrowers who are 1-15 days due in paying their loans make larger monthly payments compared to the ones whose loans have been charged off.


```python
df_sample = df_clean.sample(n=3000,replace = False)

def boxgrid(x,y, **kwargs):
    base_color = sb.color_palette()[0]
    sb.boxplot(x=x, y=y, color=base_color)

g = sb.PairGrid(data=df_sample, y_vars=['CreditScore', 'MonthlyPayment'], x_vars=['EmploymentStatus', 'LoanStatus'], height=4, aspect=1.5)
g.map(boxgrid)
g.fig.subplots_adjust(top=.9)
g.fig.suptitle('CreditScore, MonthlyPayment vs EmploymentStatus, LoanStatus')
plt.xticks(rotation=38)
plt.show();
```
