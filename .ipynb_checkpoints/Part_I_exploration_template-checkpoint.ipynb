- I'll start by filtering the columns I'll be using for the data exploration using Pandas' filter function, since I won't be needing all the 81 columns.

- The data will thereafter be cleaned for easier data exploration in this section and the others

```python
df1 = df.filter(items=['ListingNumber', 'Term', 'LoanStatus', 'BorrowerRate', 'ProsperScore', 'ListingCategory', 'EmploymentStatus', 'EmploymentStatusDuration', 'IsBorrowerHomeowner', 'CreditScoreRangeLower', 'StatedMonthlyIncome', 'TotalProsperLoans', 'LoanMonthsSinceOrigination', 'LoanOriginalAmount', 'MonthlyLoanPayment', 'TotalProsperPaymentsBilled', 'LP_CustomerPayments', 'LP_GrossPrincipalLoss'])
```

```python
df1.head(5)
```

```python
#Dimensions of the filtered dataset
df1.shape
```

```python
#Check for the sum of all the null values in the dataset
df1.isna().sum()
```

```python
#Rename some columns that have long column names to make the dataset more tidy
df1.rename(columns={'BorrowerRate':'InterestRate', 'EmploymentStatusDuration':'EmploymentDuration', 'IsBorrowerHomeOwner':'IsHomeOwner','CreditScoreRangeLower':'CreditScore', 'StatedMonthlyIncome':'MonthlyIncome', 'TotalProsperLoans':'ProsperLoans', 'LoanMonthsSinceOrigination':'LoanStretch', 'LoanOriginalAmount':'OriginalLoan', 'MonthlyLoanPayment':'MonthlyPayment', 'TotalProsperPaymentsBilled':'TotalProsperPaid'}, inplace=True)
```

```python
df1.sample(5)
```

```python
#Check the various employment statuses in the dataset
df1['EmploymentStatus'].value_counts()
```

```python
#Use the replace function to replace 'Not employed' with 'Employed'
df1.replace(to_replace='Not employed', value='Unemployed', inplace=True)
```

```python
df1['EmploymentStatus'].value_counts()
```

```python
#Instead of having 'Not available', use NaN
df1.replace('Not available', np.NaN, inplace=True)
```

```python
df1['EmploymentStatus'].value_counts()
```

```python
df1.info()
```

```python
#Get a copy of the dataset before cleaning the data for explaration
df_clean = df1.copy()
```

```python
# Drop null values
df_clean = df_clean.dropna()
```

```python
df_clean.head()
```

```python
#There's a significant change in the number of loans after dropping rows containing null values
df_clean.shape
```

```python
#Change the datatype of some columns from float to int using Pandas astype function
df_clean = df_clean.astype({'EmploymentDuration':'int64', 'ProsperLoans':'int64', 'TotalProsperPaid':'int64'})
```

## Bivariate Exploration


Looking at the correlation among the various variables in the dataset

```python
df_clean.corr()
```

```python
#Plotting a heatmap to show the correlation among the variables in an easier way to depict the linear relationships
plt.figure(figsize=[12,8])
sb.heatmap(df_clean.corr(), annot=True, fmt='.2f', cmap='vlag_r', center=0)
```

```python
#plot matrix showing the relationships between different variables
numeric_vars = ['CreditScore', 'EmploymentDuration', 'ProsperScore', 'MonthlyIncome', 'MonthlyPayment', 'OriginalLoan']
df_clean_sample = df_clean.sample(n=1000, replace=False)
g= sb.PairGrid(data=df_clean_sample, vars = numeric_vars)
g = g.map_diag(plt.hist, bins=40);
g.map_offdiag(plt.scatter)
```

As shown by the above plot matrix, Original Loan and Monthly Payment are highly correlated to each other. This could be because the amount of monthly loan payments a borrower makes highly depends on the original loan given. Also, there seems to be a positive relationship between Employment Duration and MonthlyIncome, MonthlyPayment, OriginalLoan. This shows that the longer a borrower has been employed, the higher loan given and the more the monthly payments. 
Credit Score and Employment Duration should also show a positive relationship because the longer a borrower has been employed the higher the credit score is, a further analysis will be performed. 
Prosper Score doesn't have strong correlation with the other variables, though it shows that there might be a relationship between Prosper Score and Credit Score.

```python
#Subplots showing the relationships Monthly Payment has with OriginalLoan and Monthly Income
plt.figure(figsize=[18, 6])
plt.subplot(1,2,1)
sb.regplot(data=df_clean, x='OriginalLoan', y='MonthlyPayment', x_jitter= 0.04, scatter_kws = {'alpha' : 1/10});
plt.xlabel('Original Loan')
plt.ylabel('Monthly Loan Payments');

plt.subplot(1,2,2)
sb.regplot(data=df_clean, x='MonthlyIncome', y='MonthlyPayment', x_jitter= 0.4, scatter_kws = {'alpha' : 1/5}, fit_reg=False);
plt.xlim(0, 40000)
plt.xlabel('Original Loan')
plt.ylabel('Monthly Income');
```

Original Loan is showing a linear relationship with Monthly Loan Payments. The expectation would be that the borrowers with a larger loan given pay a higher monthly payments but it appears that is not the case. There's an amount of borrowers with a smaller loan, paying more monthly payments as compared to those with larger loans.
The second plot shows that most of the loans given are given to a category of borrowers with income less than a 1000. This would explain the monthly loan payments in the first plot.

```python
#Plots showing relationships between Credit Score and ProsperScore, LP Customer Payments and Original Loan, Monthly Payments and LP Customer Payments

plt.figure(figsize=[12,14])

#subplot 1:Credit Score vs ProsperScore
plt.subplot(3,1,1)
plt.hist2d(data = df_clean, x='ProsperScore', y='CreditScore', cmin = 1, cmap = 'viridis_r')
plt.colorbar()
plt.xlabel('Prosper Scores')
plt.ylabel('Credit Scores')
#sb.regplot(data=df_clean, x='ProsperScore', y='CreditScore', fit_reg = False)

#subplot2:LP Customer Payments and Original Loan
plt.subplot(3, 1, 2)
plt.xlim(0, 26000)
sb.regplot(data=df_clean, x='OriginalLoan', y='LP_CustomerPayments', fit_reg = False, x_jitter = 0.3, scatter_kws = {'alpha' : 1/4});

#subplot3:Monthly Payments and LP Customer Payments
plt.subplot(3,1,3)
plt.xlim(0, 1300)
sb.regplot(data=df_clean, x= 'MonthlyPayment', y='LP_CustomerPayments', fit_reg = False, x_jitter = 0.4, scatter_kws = {'alpha' : 1/4});

plt.show()
```

The first subplot suggests there's no significant relationship between the Credit Scores and the Prosper Scores.
The second plot shows a linear relationship between the original loans and pre-charge off cumulative  gross payments made by the borrowers on their loans. This means that the higher the loan borrowed the larger LP_payments made by the borroower.
The third subplot suggests that there's a positive linear relationship between Monthly Loan Payments and LP_Customer payments. The trend is higher than that of Original Loans, though this may be because the monthly payments are part of the LP_Customer payments.

```python
df_sample = df_clean.sample(n=3000,replace = False)

def boxgrid(x,y, **kwargs):
    base_color = sb.color_palette()[0]
    sb.boxplot(x=x, y=y, color=base_color)
plt.figure(figsize=[8,18])
g = sb.PairGrid(data=df_sample, y_vars=['CreditScore', 'MonthlyPayment'], x_vars=['EmploymentStatus', 'LoanStatus'], height=4, aspect=1.5)
g.map(boxgrid)
g.fig.subplots_adjust(top=.9)
g.fig.suptitle('CreditScore, MonthlyPayment vs EmploymentStatus, LoanStatus')
plt.xticks(rotation=30)
plt.show();
```

Surprisingly, it appears that the unemployed borrowers have a higher credit score compared to the employed even though it's the employed who make larger monthly loan payments. Also, the retired category have a higher credit score, this is understandable because the credit may have accumulated before retirement.
Additionally, borrowers who are on their final loan payment processes are the ones with the highest credit scores. The borrowers who are 1-15 days due in paying their loans make larger monthly payments compared to the ones whose loans have been charged  off.

### Relationships observed in this part of the investigation. How did the feature(s) of interest vary with other features in the dataset?

Most of the monthly loan payments are made by borrowers with smaller loans as compared to those with larger loans. This is later explained by the incomes of those making more monthly payments having incomes less than 1000. Another interesting relationship was that of LP_CustomerPayments and Original Loans. The highest gross pre-charge off payments were made on the large loans and vice versa. Same case goes for LP_Customer Payments and Monthtly loan payments. 

Additionally, majority of the current, completed and charged off loans belong to borrowers who are either employed or self-employed. Also, there's a significant amount of borrowers who are home owners and these are the employed/full-time individuals. The retired category also have an amount of individuals who are home owners. 

There was also an unexpected relationship between Credit Scores and Employment status. Unemployed borrowers have higher credit scores compared to the individuals in the Employed category. Also, borrowers with the highest credit scores are in their final loan payment processes.

The plot suggests that majority of the Defaulted loans belong to the retired borrowers who also have the longest employment durations. Also, the unemployed borrowers have a high number of charged off loans with short employment durations.

According to the plots, the highest percentage of the loans are the recently acquired loans/ Current which belong to majority of the employed individuals with some receiving loans of upto 35000 and earning salaries of even upto 40000. The loans that are charged off mostly are from low income earners with <20000 but at the same time receiving high loans.

## Conclusions
- Majority of the loans are the current loans, not yet reached their due date and most of these belong to the employed borrowers.
